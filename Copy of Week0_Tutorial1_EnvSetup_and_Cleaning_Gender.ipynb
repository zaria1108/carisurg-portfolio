### 2.1 Variables and Data Types

```python
# Variables store values. Python figures out the type automatically.
patient_id    = 42          # integer
patient_name  = "Maria"     # string
heart_rate    = 88.5        # float
is_critical   = True        # boolean

print(f"Patient:    {patient_name}")
print(f"Heart rate: {heart_rate} bpm")
print(f"Critical:   {is_critical}")
print(f"Type of heart_rate: {type(heart_rate)}")
```

### 2.2 Lists

```python
# A list holds multiple values in order
vital_signs = ["pulse", "temperature", "SBP", "DBP", "RR", "FiO2", "GCS"]

print("All vitals:", vital_signs)
print("First vital:", vital_signs[0])      # indexing starts at 0
print("Last vital:", vital_signs[-1])      # -1 gets the last item
print("Number of vitals:", len(vital_signs))
```

### 2.3 Loops

```python
# A for loop runs a block of code once for each item in a list
for vital in vital_signs:
    print(f"  - {vital}")
```

### 2.4 Conditional Logic (if / elif / else)

```python
# Conditionals let code make decisions
def classify_pulse(bpm):
    if bpm < 60:
        return "Bradycardia (too slow)"
    elif bpm > 100:
        return "Tachycardia (too fast)"
    else:
        return "Normal"

# Test it
test_pulses = [45, 78, 130]
for p in test_pulses:
    print(f"  Pulse {p} bpm -> {classify_pulse(p)}")
```

### 2.5 Functions

```python
# Functions package reusable code
def greet_trainee(name, unit="Clinical AI & Innovation Unit"):
    message = f"Welcome to Mercer General, {name}. You are now part of the {unit}."
    return message

print(greet_trainee("Maria")) # If the unit is not specified it is reverts to the default variable in its definition.
print(greet_trainee("James", "Emergency Department"))
```

## 4. Day 1 Challenge — Cleaning the Gender Column

**[TUTOR NOTE — this is the core of Tutorial 1, ~25 min]**

**First, explore before cleaning.** Never clean data you have not looked at.
Walk through each step below and explain the reasoning out loud.


```python
# Step 1: What values are actually in the Gender column?
print("Unique values in Gender:")
print(df_raw['Gender'].unique())
print(f"\nTotal unique values: {df_raw['Gender'].nunique()}")
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

**[TUTOR NOTE — discussion pause]**
You should see six variants: `0`, `1`, `Male`, `MALE`, `Female`, `FEMALE`.

Ask the class: *Should we represent Gender as M/F or 1/0? What are the trade-offs?*

Key points to draw out:
- `1/0` (integer encoding) is easier for models to process later
- `M/F` is more readable for humans
- The programme uses `1 = Male, 0 = Female` — consistent with the Harvard dataset
- This is a *design decision* — what matters is that you document it


```python
# Step 2: Count how many of each value we have
print("Gender value counts:")
print(df_raw['Gender'].value_counts())
```

```python
# Step 3: Build the mapping — every variant maps to 1 or 0
gender_map = {
    'Male':   1,
    'MALE':   1,
    '1':      1,
    'Female': 0,
    'FEMALE': 0,
    '0':      0
}

# Step 4: Apply the mapping to create a new, clean column
# It is best practice to make changes to a new column rather than editing an
# exising one.
df_raw['Gender_Clean'] = df_raw['Gender'].map(gender_map)

# Step 5: Verify — check what we have now
print("After cleaning:")
print(df_raw['Gender_Clean'].value_counts())
print(f"\nAny NaN values? {df_raw['Gender_Clean'].isnull().sum()}")
```

```python
# Step 6: Drop the original dirty column now that we have a clean one
df_raw = df_raw.drop(columns=['Gender'])
df_raw = df_raw.rename(columns={'Gender_Clean': 'Gender'})

# Confirm the result
print("Column 'Gender' after cleaning:")
print(df_raw['Gender'].value_counts())
print(f"Data type: {df_raw['Gender'].dtype}")
df_raw.head()
```

**[TUTOR NOTE — common errors to watch for]**

1. **Using `.replace()` instead of `.map()`** — both work, but `.map()` is safer for column-wide remapping because any unmatched value becomes `NaN` (a useful warning signal).
2. **Forgetting `.inplace=True` or reassignment** — a very common beginner mistake. Always show students `df['col'] = df['col'].something()` vs `df['col'].something(inplace=True)`.
3. **Case sensitivity** — `'male'` and `'Male'` are different strings. Show them `str.lower()` as an alternative approach if the mapping gets unwieldy.

**Alternative approach (bonus, if time permits):**


```python
# Alternative: normalise to lowercase first, then map
# This handles any future casing variants automatically
gender_map_v2 = {'male': 1, 'female': 0, '1': 1, '0': 0}

# Reset and try the alternative approach on a copy
df_test = pd.read_csv(FILE_PATH)
df_test['Gender_v2'] = df_test['Gender'].astype(str).str.lower().str.strip().map(gender_map_v2)
print(df_test['Gender_v2'].value_counts())
print(f"NaNs: {df_test['Gender_v2'].isnull().sum()}")
```
